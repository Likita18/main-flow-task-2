```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Global_Superstore2.csv", encoding='ISO-8859-1')  # adjust encoding if needed

# Initial inspection
print(df.head())
print(df.info())
print(df.describe())

```

```python
# Check missing values
print(df.isnull().sum())

# Option 1: Fill missing values (numerical only)
df.fillna(df.mean(numeric_only=True), inplace=True)

# Option 2: Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Apply on key numerical columns
for col in ['Sales', 'Profit']:
    df = remove_outliers_iqr(df, col)
```

```python
# Summary statistics
print(df[['Sales', 'Profit']].describe())

# Correlation matrix
corr_matrix = df[['Sales', 'Profit']].corr()
print(corr_matrix)
```

```python
df[['Sales', 'Profit']].hist(bins=30, figsize=(10, 4))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Sales', 'Profit']])
plt.title('Boxplot of Sales and Profit')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Sales', 'Profit']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum, ci=None)
plt.title('Total Sales by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum, ci=None)
plt.title('Total Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
df.to_csv("Cleaned_Global_Superstore.csv", index=False)
```
